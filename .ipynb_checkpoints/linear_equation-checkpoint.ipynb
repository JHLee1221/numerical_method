```python
import numpy as np
import scipy as scp

%matplotlib widget
import matplotlib.pyplot as plt
import mpl_toolkits.mplot3d as plt3d
```

```python
# assigning the matrix A
A = np.array([ [1, 3, -2, 5, -6], 
               [3, 5, 6, 10, 1], 
               [2, 4, 3, -2, 9], 
               [8, -2, 3, -1, 4], 
               [1, 4, -10, -2, -6] ])

# for debugging
print(A)
```

```python
# assigning the vector B
b = np.array([[5, 7, -2, 10, -8]])
b = b.T

# for debugging
print(b)
```

```python
# solving the linear equation, basic numpy/scipy feature
Xnp = np.linalg.solve(A, b)
Xsp = scp.linalg.solve(A, b)

# for debugging
print("Xnp=", Xnp)
print("Xsp=", Xsp)
```

```python
# calculating LU decomposition matrix
P, L, U = scp.linalg.lu(A)

# for debugging
print("L=", L)
print("U=", U)
print("P=", P)
```

```python
# solving the linear equation usint the LU decomposition
LU, P = scp.linalg.lu_factor(A)
XspLU = scp.linalg.lu_solve((LU, P), b)

# for debugging
print("LU=", LU)
print("P=", P)
print("XspLU=", XspLU)
```

```python
# calculating inverse matrix
Ainv = np.linalg.inv(A)

# for debugging
print(Ainv)
```

```python
# solving the linear equation, using the inverse matrix
XnpInv = np.matmul(Ainv, b)

# for debugging
print("XnpInv=", XnpInv)
```

```python
# solving the linear equation, usint the iterative method
XspIter = scp.sparse.linalg.lsmr(A, b)[0]

# for debugging
print("XspIter=", XspIter)
```

```python
gmin = -20
gmax = 20
samp = 50
geo = np.linspace(gmin, gmax, samp)

X, Y = np.meshgrid(geo, geo)

fig = plt.figure(figsize = (8, 8))
ax = fig.add_subplot(111, projection  = '3d')
Z1 = 1/2 - (1/2)*X - (3/2)*Y
Z2 = 2/3 + (2/3)*X - (1/3)*Y
Z3 = 3 - 3*X + 5*Y
ax.plot_surface(X, Y, Z1, color = 'r')
ax.plot_surface(X, Y, Z2, color = 'g')
ax.plot_surface(X, Y, Z3, color = 'b')
plt.show()
```

```python
def gauss_jordan(matrix):
    A = np.array(matrix, dtype=float)

    n = len(A)
    for i in range(n):
        # 선택된 행의 첫 번째 요소를 1로 만들기 위한 비율 계산
        scale = A[i][i]
        A[i] = A[i] / scale

        # 선택된 행을 사용하여 다른 모든 행의 첫 번째 요소를 0으로 만듦
        for j in range(n):
            if i == j:
                continue
            A[j] = A[j] - A[i] * A[j][i]

    return A

if __name__ == "__main__":
    # 주어진 선형 연립방정식의 확장된 행렬
    # augmented_matrix = [
    #     [2, -1, 5, 1],
    #     [3, 2, 2, -6],
    #     [1, 3, 3, -1]
    # ]
    
    augmented_matrix = np.array([ [1, 3, -2, 5, -6, 5], 
               [3, 5, 6, 10, 1, 7], 
               [2, 4, 3, -2, 9, -2], 
               [8, -2, 3, -1, 4, -10], 
               [1, 4, -10, -2, -6, -8] ])    

    reduced_matrix = gauss_jordan(augmented_matrix)

    print("Gauss-Jordan 알고리즘을 적용한 확장된 행렬:")
    print(reduced_matrix)

    # 결과 출력
    variables = ['x', 'y', 'z', 'ㅏ']
    for i, var in enumerate(variables):
        print(f"{var} = {reduced_matrix[i][-1]}")
```
